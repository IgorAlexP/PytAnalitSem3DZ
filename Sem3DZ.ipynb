## 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома

```python
import pandas as pd

# Чтение CSV-файла
data = pd.read_csv('kc_house_data.csv')

# Создание нового признака delta_renovated
data['delta_renovated'] = data['yr_renovated'] - data['yr_built']

# Замена значений NaN в delta_renovated на 0
data['delta_renovated'].fillna(0, inplace=True)

# Вывод первых 5 записей для проверки
print(data.head())
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
import pandas as pd

# Чтение CSV-файла
data = pd.read_csv('kc_house_data.csv')

# Преобразование столбца 'date' в тип данных datetime
data['date'] = pd.to_datetime(data['date'])

# Создание признака 'year_sold' - год продажи
data['year_sold'] = data['date'].dt.year

# Создание признака 'month_sold' - месяц продажи
data['month_sold'] = data['date'].dt.month

# Вывод первых 5 записей для проверки
print(data.head())
```

## 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
import pandas as pd

# Создание DataFrame с данными о домах
houses = pd.DataFrame({
    'bedrooms': [3, 4, 2, 5, 3, 4, 2, 3, 5, 4, 3, 2, 4, 3, 2, 4, 3, 5, 4, 3, 2, 4,
                 3, 5, 2, 4, 3, 5, 2, 4, 3, 5],
    'price': [500000, 600000, 400000, 700000, 450000, 550000, 380000, 420000,
              650000, 570000, 490000, 360000, 530000, 480000, 300000, 550000,
              420000, 680000, 520000, 450000, 390000, 610000, 530000, 690000,
              410000, 620000, 550000, 700000, 480000, 460000, 580000, 510000]
})

# Нахождение средней стоимости домов в зависимости от количества спален
average_price_by_bedrooms = houses.groupby('bedrooms')['price'].mean()

# Сортировка по возрастанию стоимости
sorted_average_price = average_price_by_bedrooms.sort_values()

# Вывод результата
print(sorted_average_price)
```

## 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
# Группировка данных по состоянию дома и вычисление статистик
house_prices_by_condition = data.groupby('condition')['price'].agg(['min', 'mean', 'max'])

# Вывод результатов
print(house_prices_by_condition)
```

## 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
# Построение таблицы с подсчетом количества домов
table = data.pivot_table(index='waterfront', columns='view', values='id', aggfunc='count')

# Вывод таблицы
print(table)
```

## 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
# Построение таблицы с подсчетом количества домов
table = pd.crosstab(index=data['floors'], columns=data['bedrooms'])

# Вывод таблицы
print(table)
```

## 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
# Построение таблицы с подсчетом медианной стоимости домов
table = data.pivot_table(values='price', index='condition', columns='grade', aggfunc='median')

# Вывод таблицы
print(table)
```
